```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv('cancerdata.csv')
dataset.head()
```

```python
# splitting X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataset,dataset['diagnosis'], test_size=0.25,random_state=2)
```

```python
X_train.drop(['diagnosis','id'],axis=1,inplace=True)
X_test.drop(['diagnosis','id'],axis=1,inplace=True)
```

```python
X_train.head()
```

Since KNN relies on a distance measure to figure out which class a test observation belongs to. Scaling the data becomes a must here since if we do not scale the data then features with larger values will be dominant.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

```python
# comparing actual response values (y_test) with predicted response values (y_pred)
from sklearn import metrics
print("KNN model accuracy:", metrics.accuracy_score(y_test, y_pred))
```

 #### for K -Calculation

```python
from sklearn.model_selection import GridSearchCV

#create new a knn model
knn2 = KNeighborsClassifier()

#create a dictionary of all values we want to test for n_neighbors
param_grid = {'n_neighbors': np.arange(1, 25)}

#use gridsearch to test all values for n_neighbors
knn_gscv = GridSearchCV(knn2, param_grid, cv=10)

#fit model to data
knn_gscv.fit(X_train, y_train)
```

```python
#check top performing n_neighbors value
knn_gscv.best_params_
```

```python
#check mean score for the top performing value of n_neighbors
knn_gscv.best_score_
```

```python
classifier = KNeighborsClassifier(n_neighbors=7)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

```python
print("KNN model accuracy:", metrics.accuracy_score(y_test, y_pred))
```
